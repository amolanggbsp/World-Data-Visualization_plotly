### Age Structure Chorpleth Plot (world map)
##### 1. 65+ aged population proportion (pink color) 
##### 2. age 15-24 proportion (blue color)
##### 3. age 25-54 proportion (green color)

```python
import pandas as pd
age_structure = pd.read_csv("age_structure.csv") 
age_structure
```

## 1. 65+ aged population proportion (pink color) 

```python

#set color as pink 
scl = [
    [0.0, 'rgb(20, 0, 14)'],
    [0.2, 'rgb(79, 7, 61)'],
    [0.4, 'rgb(206, 30, 153)'],
    [0.6, 'rgb(239, 93, 195)'],
    [0.8, 'rgb(239, 184, 223)'],
    [1.0, 'rgb(252, 244, 250)']
]



#set module type, colorscale, marker option, text, label option, color bar
meta_data = dict(type = 'choropleth', 
           locations = age_structure['country name'], 
           colorscale = scl, 
           reversescale = True, 
           locationmode = 'country names', 
           z = age_structure['65+'], 
           marker= dict(line=dict(color='black', 
                        width=1)), 
           text = age_structure['country name'], 
           hoverlabel= dict(bgcolor= '#e209ac',
                            font=dict(family='Times New Roman',
                                      color='white')), 
           colorbar = {'title':'Aged Rate', 'nticks':9}
           )   

#set layout: title, geographic range 
layout = dict(title = 'Global Choropleth Plot By Aged Rate(65+%)', 
             geo = dict(showframe = True, 
                       projection = {'type': 'equirectangular'})) 

#use my_choromap module for worldmap 
My_choromap = go.Figure(data = [meta_data], layout=layout)
py.iplot(My_choromap)
```
