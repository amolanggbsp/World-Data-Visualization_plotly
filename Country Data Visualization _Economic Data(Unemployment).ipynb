```python
import csv
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go

#2007 data visualization 
unemployment = pd.read_csv('unemployment.csv')
unemployment
country = unemployment['country name']
data2007 = unemployment['2007']

#set: x axis data, y axis data, line color and width 
trace0 = go.Scatter(
    x = country,
    y = data2007,
    name = '2007',
    line = dict(
        color = ('rgb(205, 12, 24)'),
        width = 4)
)
data = [trace0]

#set layout: title, xaxis, yaxis 
layout = dict(title = 'Unemployment Rate Change',
              xaxis = dict(title = 'country name'),
              yaxis = dict(title = 'unemploytment %'),
              )

fig = dict(data=data, layout=layout)
py.iplot(fig, filename='styled-line')
```

### Unemployment rate change graph (2006-2014/  9 countries)

```python
import csv
import pandas as pd

import plotly.plotly as py
import plotly.graph_objs as go

unemployment = pd.read_csv('unemployment.csv')
unemployment1 = unemployment.T
unemployment1.iloc[ : , 3]
year = ['2006','2007','2008','2009','2010','2011','2012','2013','2014']  #x-axis data 
country= unemployment1.iloc[ 0 , : ]
unemployment2 = unemployment1.iloc[1:,:]
unemployment2
spain = unemployment2.iloc[ : , 0]
philippines= unemployment2.iloc[ : , 1]
south_korea=unemployment2.iloc[ : , 2]
mexico=unemployment2.iloc[ : , 3]
austria=unemployment2.iloc[ : , 4]
poland=unemployment2.iloc[ : , 5]
portugal=unemployment2.iloc[ : , 6]
egypt=unemployment2.iloc[ : , 7]
denmark=unemployment2.iloc[ : , 8]
australia=unemployment2.iloc[ : , 9]

#each trace = each country's unemployment data 

#trace: xaxis, yaxis, line color, lind width 
trace0 = go.Scatter(
    x = year,
    y = spain,
    name = 'spain',
    line = dict(
        color = ('rgb(205, 12, 24)'),
        width = 4)
)

trace1 = go.Scatter(
    x = year,
    y = philippines,
    name = 'philippines',
    line = dict(
        color = ('rgb(400, 200, 24)'),
        width = 4)
)
trace2 = go.Scatter(
    x = year,
    y = south_korea,
    name = 'south_korea',
    line = dict(
        color = ('rgb(800, 200, 124)'),
        width = 4)
)
trace3 = go.Scatter(
    x = year,
    y = mexico,
    name = 'mexico',
    line = dict(
        color = ('rgb(20, 600, 4)'),
        width = 4)
)

trace4 = go.Scatter(
    x = year,
    y = austria,
    name = 'austria',
    line = dict(
        color = ('rgb(500, 10, 44)'),
        width = 4)
)

trace5 = go.Scatter(
    x = year,
    y = poland,
    name = 'poland',
    line = dict(
        color = ('rgb(900,2,400)'),
        width = 4)
)

trace6 = go.Scatter(
    x = year,
    y = portugal,
    name = 'portugal',
    line = dict(
        color = ('rgb(200, 800, 400)'),
        width = 4)
)

trace7 = go.Scatter(
    x = year,
    y = egypt,
    name = 'egypt',
    line = dict(
        color = ('rgb(2, 600, 400)'),
        width = 4)
)

trace8 = go.Scatter(
    x = year,
    y = denmark,
    name = 'denmark',
    line = dict(
        color = ('rgb(2, 100, 4)'),
        width = 4)
)

data = [trace0, trace1,trace2, trace3, trace4, trace5, trace6, trace7, trace8]
#set layout: title, xaxis, yaxis 
layout = dict(title = 'Unemployment Rate Change',
              xaxis = dict(title = 'Year'),
              yaxis = dict(title = 'Unemployment %'),
    
              )

fig = dict(data=data, layout=layout)
py.iplot(fig, filename='legend-names')
```
