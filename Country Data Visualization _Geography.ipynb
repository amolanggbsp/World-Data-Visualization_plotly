# Geographic Data Visualization 
### 1. land_use data 
### 2. land_use data(agriculture%) + area_composition data(land %) subplot

##### * agriculture % : proportion of agriculture land       *land %: proportion of land of total country size

```python
import pandas as pd

land_use = pd.read_csv('land_use.csv')
area_comp = pd.read_csv('area_comp.csv')
land_use1= land_use.drop(land_use.index[[5,22,26]])   #delete 3 lines 
land_use2= land_use1.sort_values(by='agriculture_land', ascending = False)  #sort values by 'argiculture' size 
land_use2
```

```python

area_comp1= area_comp.drop(area_comp.index[[5,22,26]]) #delete 3 lines 
area_comp1
```

#### Add necessary columns (land %, water%) 

```python
area_comp1['land%'] = area_comp1['land'].div(area_comp1['total'])*100
area_comp1['water%'] = area_comp1['water'].div(area_comp1['total'])*100

area_comp2 = area_comp1.sort_values(by='land%', ascending = False)
area_comp2
```

### Horizontal Bar Graph 
##### show proportion of each component(argriculture, forest, etc) 
###### agriculture: pink,    forest: black    , etc: green

```python
import plotly.plotly as py
import plotly.graph_objs as go
land_use1= land_use.drop(land_use.index[[5,22]])
land_use1

#each trace : proportion of each component 
#trace: x axis data, y axis data, set horizontal, marker option, line option 
trace1 = go.Bar(
    y=land_use2['country name'],
    x=land_use2['agriculture_land'],
    name='agriculture',
    orientation = 'h',
    marker = dict(
        color = 'rgba(246, 78, 139, 0.6)',
        line = dict(
            color = 'rgba(246, 78, 139, 1.0)',
            width = 2)
    )
)
trace2 = go.Bar(
    y=land_use2['country name'],
    x=land_use2['forest'],
    name='forest ',
    orientation = 'h',
    marker = dict(
        color = 'rgba(58, 71, 80, 0.6)',
        line = dict(
            color = 'rgba(58, 71, 80, 1.0)',
            width = 2)
    )
)

trace3 = go.Bar(
    y=land_use2['country name'],
    x=land_use2['other'],
    name='other',
    orientation = 'h',
    marker = dict(
        color = 'rgba(50, 171, 190, 0.9)',
        line = dict(
            color = 'rgba(50, 171, 190, 1.0)',
            width = 2)
    )
)

data = [trace1, trace2, trace3]

#set layout: width, height, title, xaxis name
layout = go.Layout(    autosize=False,
    width=1000,
    height=900,
                   title = 'Land Use',
              xaxis = dict(title = 'Percentage % '), barmode='stack'
    
              )

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='marker-h-bar')
```
